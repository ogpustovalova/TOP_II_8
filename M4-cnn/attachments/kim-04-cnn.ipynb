## Часть А. Свёртка с нуля

Операция двумерной свёртки:

$$(I \star K)[i,j] = \sum_m \sum_n I[i+m, j+n] \cdot K[m, n]$$

Размер выхода при шаге $S$ и дополнении $P$:

$$H_{out} = \left\lfloor \frac{H_{in} - K + 2P}{S} \right\rfloor + 1$$
