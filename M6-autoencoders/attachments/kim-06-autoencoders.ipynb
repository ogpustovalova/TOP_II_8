## Часть В. Вариационный автокодировщик (VAE) и генерация

VAE предполагает, что латентное представление распределено нормально: энкодер
выдаёт $\mu$ и $\log \sigma^2$, а сэмплы $z$ берутся через репараметризацию:
$$z = \mu + \sigma \odot \varepsilon, \quad \varepsilon \sim \mathcal{N}(0, I)$$

Loss VAE = реконструкция + KL-дивергенция:
$$D_{KL} = -\frac{1}{2} \sum_i \bigl(1 + \log \sigma_i^2 - \mu_i^2 - \sigma_i^2\bigr)$$
